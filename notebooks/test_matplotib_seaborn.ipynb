```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Dữ liệu
categories = ['A', 'B', 'C', 'D']
values = [10, 25, 15, 30]

# Tạo biểu đồ
fig, ax = plt.subplots()
bars = ax.bar(categories, values)  # Tạo container

# Kiểm tra containers
print(ax.containers)  # [<BarContainer object of 4 artists>]
print(len(ax.containers))  # 1

# Thêm nhãn
ax.bar_label(ax.containers[0])  # Thêm label cho cột đầu tiên (cũng là duy nhất)

plt.show()
```
